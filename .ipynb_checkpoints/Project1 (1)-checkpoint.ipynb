```python
query = "SELECT restaurant_data.camis as rest_camis,violation_data.camis as viol_camis, restaurant_data.boro, restaurant_data.zipcode, violation_data.grade,violation_data.critical_flag FROM restaurant_data INNER JOIN violation_data ON restaurant_data.camis = violation_data.camis;"
joined_df = pd.read_sql(query, sql_conn)
```
